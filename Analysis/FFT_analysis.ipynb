```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
filename = '220622test.xlsx'
df = pd.read_excel(filename, usecols='A,T:Y', engine='openpyxl')
```

```python
TIME = df['time'].tolist()
SQRT_ACC_1 = df['SQRT_ACC_1'].tolist()
SQRT_ANG_1 = df['SQRT_ANG_1'].tolist()
SQRT_AV_1 = df['SQRT_AV_1'].tolist()
SQRT_ACC_2 = df['SQRT_ACC_2'].tolist()
SQRT_ANG_2 = df['SQRT_ANG_2'].tolist()
SQRT_AV_2 = df['SQRT_AV_2'].tolist()
```

```python
plt.figure(figsize=(30,18))
plt.subplot(3,1,1)
plt.plot(TIME, SQRT_ACC_1)
plt.grid(True)
plt.subplot(3,1,2)
plt.plot(TIME, SQRT_ANG_1)
plt.grid(True)
plt.subplot(3,1,3)
plt.plot(TIME, SQRT_AV_1)
plt.grid(True)
plt.show
```

```python
L = len(SQRT_ACC_1)
fft=[[0]*L]*3
fft_magnitude = [[0]*L]*3
fft[0] = np.fft.fft(SQRT_ACC_1) / L
fft[1] = np.fft.fft(SQRT_ANG_1) / L
fft[2] = np.fft.fft(SQRT_AV_1) / L
for i in range(3):
  fft_magnitude[i] = abs(fft[i])
```

```python
plt.figure(figsize=(30,18))
plt.subplot(3,1,1)
plt.plot(fft_magnitude[0])
plt.grid(True)
plt.subplot(3,1,2)
plt.plot(fft_magnitude[1])
plt.grid(True)
plt.subplot(3,1,3)
plt.plot(fft_magnitude[2])
plt.grid(True)
plt.show
```

```python
plt.figure(figsize=(30,18))
plt.subplot(3,1,1)
plt.plot(TIME, SQRT_ACC_2)
plt.grid(True)
plt.subplot(3,1,2)
plt.plot(TIME, SQRT_ANG_2)
plt.grid(True)
plt.subplot(3,1,3)
plt.plot(TIME, SQRT_AV_2)
plt.grid(True)
plt.show
```
